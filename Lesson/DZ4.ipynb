```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
plt.plot(t, f, color = 'green')
plt.title("График f(t)")
plt.xlabel("Значение t")
plt.ylabel("Значение f")
plt.axis([0.5, 9.5, -2.5, 2.5])
```
